# **Задание**

Построить сводные таблицы на основе данных, хранящихся в файле accidents.

Импортируйте библиотеки  NumPy и Pandas

```python
# Ваш код здесь
import pandas as pd
import numpy as np
```

Считайте данные из файла `accidents.xlsx`, который содержит столбцы:

* Код региона
* Название региона
* Вид дороги
* Адрес
* Вид ДТП
* Дата ДТП
* Время ДТП
* Причина ДТП
* Количество погибших
* Количество пострадавших
* Количество ТС
* Количество участников



```python
# Ваш код здесь
accidents = pd.DataFrame(pd.read_excel('accidents.xlsx'))
accidents
```

Создайте простую сводную таблицу. В качестве индекса выберете столбец "Название региона".

```python
# Ваш код здесь
pivot_table = pd.pivot_table(accidents, index = ['Название региона'], aggfunc='count')
```

Добавьте колонки с Видом ДТП

```python
# Ваш код здесь
pd.pivot_table(accidents,index=['Название региона'],columns=['Вид ДТП'], aggfunc='count')
```

Теперь у нас данные организованы по строкам и столбцам и содержат информацию о разных городах и видах произошедших ДТП

Оставим в сводной таблице данные только о количестве транспортных средств, участвовавших в ДТП.

```python
# Ваш код здесь
pd.pivot_table(accidents,index=['Название региона'],columns=['Вид ДТП'], values =['Количество ТС'])
```

Модернизируйте сводную таблицу так, чтобы выполнялось два уровня группировки:

1. Вид дороги
2. Название региона

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Вид дороги', 'Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС'],
               aggfunc='mean')
```

Модернизируйте сводную таблицу таким образом, что в значениях было посчитано не среднее количество ТС, а суммарное.

Приведите полученные данные к целочисленному типу

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС'], 
               aggfunc='sum').fillna(0).astype(int) 
```

Очистите сводную таблицу от значений `NaN`. Замените их на `0`.

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС'], 
               aggfunc='sum',
               fill_value=0)
```

Добавьте количество пострадавших в список значений `values`:

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС', 'Количество пострадавших'], 
               aggfunc='sum',
               fill_value=0)
```

Дополните сводную таблицу строкой ИТОГО

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС', 'Количество пострадавших'], 
               aggfunc='sum',
               fill_value=0,
              margins=True)
```

Посчитатйте в сводной таблице суммарные значения для Количества ТС и средние для количества пострадавших

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС', 'Количество пострадавших'], 
               aggfunc=['sum', 'mean'],
               fill_value=0,
              margins=True)
```

Дополните сводную таблицу данными с количеством произошедших ДТП

```python
# Ваш код здесь
pd.pivot_table(accidents,
               index=['Название региона'],
               columns=['Вид ДТП'], 
               values =['Количество ТС', 'Количество пострадавших'], 
               aggfunc=['sum', 'mean', len],
               fill_value=0,
              margins=True)
```

Представьте столбец `Вид дороги` как категориальную переменную



```python
# Ваш код здесь
accidents["Вид дороги"] = accidents["Вид дороги"].astype("category")
```
